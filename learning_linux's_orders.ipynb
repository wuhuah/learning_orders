本文学习Linux的相关命令


切换至根目录

```
cd /
```

切换至绝对路径

```
cd /content/drive/My Drive/Face-mask-detector-using-RetinaNet-model/keras-retinanet
```

切换至上级目录

```
cd ../
```

查看当前路径下的文件和文件夹名称

```
ls
```

显示当前路径下的所有文件及文件夹的详细信息。

```
ls -1
```

显示指定路径下的所有文件及文件夹的详细信息。

```
ls -1 keras-retinanet/
```

查看根目录下的文件和文件夹名称

```
ls /
```

```
ls *1 wc
```

读取指定文件内容，可相对或绝对路径

```
cat keras-retinanet/requirements.txt
```

查看当前命令所在路径

```
!pwd
```

删除文件或文件夹

```
rm -r ./need_delete1
```

仅可删除文件，不可删文件夹

```
rm -f ./need_delete.ipynb
```

创建单级文件夹

```
mkdir ./newdir
```

创建多级文件夹

```
mkdir -p ./newdir/a/b/c
```

复制当前路径下的文件到另一个路径下

```
cp ./maskDetectorData.csv ./snapshots
```

复制当一个径下的文件夹到另一个路径下

```
cp -r ./snapshots newdir/a/b
```
